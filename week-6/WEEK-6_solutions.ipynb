```python
#9 prepare X and Y
x =  X.iloc[:,1:].values
y =  X.iloc[:,0].values
print(x.shape)
print(y.shape)
```

```python
#alternate way
p = X.drop(columns=["label"])
q = X["label"]
print(p.shape)
print(q.shape)
```

```python
#10.visualization the image of the first 5 digital in x also print class label in 
plt.figure(figsize=(10, 5))

for i in range(5):
    plt.subplot(1, 5, i + 1)  # Create a subplot (1 row, 5 columns)
    plt.imshow(x[i].reshape(28, 28),cmap='gray')  # Reshape 784 to 28x28
    plt.title(f"Label: {y[i]}")  # Show class label
    plt.axis("off")  # Hide axis

plt.show()
```

```python
#11.split the data set into training set(75%) and validation set(25%)
from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val =train_test_split(x, y, test_size= 0.25, random_state=42)

print(X_train.shape, y_train.shape)
print(X_val.shape, y_val.shape)
```

```python
#12 & 13.fit KNN clasifies the train data
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors = 5, metric = "euclidean",p = 2)

knn.fit(X_train, y_train)
```

```python
#14.predict the validation set results
y_predict = knn.predict(X_val)
```

```python
#15print the confusion matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_predict, y_val)
print("confusion matrix")
print(cm)
```

```python
#16.print classification report and accuracy
from sklearn.metrics import classification_report, accuracy_score

accuracy = accuracy_score(y_val, y_predict)
cr = classification_report(y_val, y_predict)
print("accuracy")
print(accuracy)
print("\n")
print("classification report")
print(cr)
```

```python
#17.visualize the image of the first five digits in X-test and print the predicted

plt.figure(figsize=(10, 5))

for i in range(10):
    plt.subplot(2, 5, i + 1)  
    plt.imshow(X_val[i].reshape(28, 28), cmap="gray") 
    plt.title(f"Pred: {y_predict[i]}") 
    plt.axis("off") 

plt.show()
```
