```python
#19.prepare x and y
X = df.drop(columns=['Species'])
y = df['Species']
```

```python
#20.split into train and test data 70% train random state = 1
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

```python
#21.Look at the documentation of the SVC class then you will use to create the model
from sklearn.svm import SVC
```

```python
#22.model building
sv = SVC()#You're using default SVC settings: kernel='rbf', C=1.0, gamma='scale'

model = sv.fit(X_train, y_train);
```

```python
#23.Instantiate an object of class SCV using cost c =1
sv = SVC(C=1)
```

```python
#24.Fit the model
model = sv.fit(X_train, y_train)
```

```python
#25.Predict using the model
y_pred = model.predict(X_test)
```

```python
#26.Evalution the model using confusion matrix
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n", cm)
```

```python
#27.print other metrics
from sklearn.metrics import classification_report, accuracy_score

# Print classification report
print("Classification Report:\n", classification_report(y_test, y_pred))

# Print overall accuracy
print("Accuracy Score:", accuracy_score(y_test, y_pred))
```

```python
#28.print classification report ans accuracy
from sklearn.metrics import classification_report, accuracy_score

# Classification report
print("Classification Report:\n", classification_report(y_test, y_pred))

# Accuracy
print("Accuracy Score:", accuracy_score(y_test, y_pred))
```

```python
#29.K-fold cross validation
from sklearn.model_selection import cross_val_score

# Create the model again (optional but clearer)
sv = SVC(C=1)

# Perform 5-fold cross-validation (you can change cv=5 to cv=10, etc.)
scores = cross_val_score(sv, X, y, cv=5)

# Print individual scores and average
print("Cross-validation scores:", scores)
print("Average accuracy:", scores.mean())
```

```python
#30.use grid search CV to find optimal value of c-sel decision tree lab questions
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC

# Define the parameter grid
param_grid = {'C': [0.1, 1, 10, 100, 1000]}

# Set up the GridSearchCV
grid = GridSearchCV(SVC(), param_grid, cv=5)

# Fit the model on the entire dataset
grid.fit(X, y)

# Print best parameters and score
print("Best Parameters:", grid.best_params_)
print("Best Cross-Validation Score:", grid.best_score_)
```

```python
#31.Try with C=[0,1,1,10,100,1000]
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC

grid = GridSearchCV(SVC(), {'C': [1, 10, 100, 1000]}, cv=5)
grid.fit(X, y)
print("Best C:", grid.best_params_['C'])
print("Best Score:", grid.best_score_)
```

```python
#32.specify no of folds for K-fold CV
from sklearn.model_selection import KFold, cross_val_score
from sklearn.svm import SVC

kf = KFold(n_splits=5, shuffle=True, random_state=1)  # 5-fold CV
model = SVC(C=1)
scores = cross_val_score(model, X, y, cv=kf)

print("CV Scores:", scores)
print("Average Score:", scores.mean())
```

```python
#35.plot of c verrus train and test scores
import matplotlib.pyplot as plt
grid = GridSearchCV(SVC(), {'C': [1, 10, 100, 1000]}, cv=5, return_train_score=True)
grid.fit(X, y)
C_values = results_df['param_C'].astype(float)
train_scores = results_df['mean_train_score']
test_scores = results_df['mean_test_score']

plt.plot(C_values, train_scores, label='Train Score', marker='o')
plt.plot(C_values, test_scores, label='Test Score', marker='s')
plt.xlabel('C value')
plt.ylabel('Score')
plt.title('C vs Train/Test Scores')
plt.legend()
plt.grid(True)
plt.show()
```
